```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv("Table_4.5.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Reliance -Jio"] = df["Reliance -Jio"].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
df.rename(columns={df.columns[0]: "Year"}, inplace=True)
```

```python
df["Year"] = df["Year"].astype(int)
```

```python
df.set_index("Year", inplace=True)
```

```python
plt.figure(figsize=(12, 6))
for column in df.columns:
    sns.lineplot(x=df.index, y=df[column], label=column)

plt.xlabel("Year")
plt.ylabel("Subscribers (in millions)")
plt.title("Telecom Subscriber Growth Over Years")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.sum().sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.xlabel("Telecom Company")
plt.ylabel("Total Subscribers (in millions)")
plt.title("Total Subscribers per Telecom Company")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Between Telecom Companies")
plt.show()
```

```python
year_to_plot = 2015
if year_to_plot in df.index:
    df.loc[year_to_plot].plot(kind='bar', figsize=(12, 6), color='coral')
    plt.xlabel("Telecom Company")
    plt.ylabel("Subscribers (in millions)")
    plt.title(f"Subscriber Distribution in {year_to_plot}")
    plt.xticks(rotation=45)
    plt.show()
else:
    print(f"Year {year_to_plot} not found in dataset.")
```

```python
plt.figure(figsize=(12, 6))
for column in df.columns:
    sns.lineplot(x=df.index, y=df[column].rolling(window=3).mean(), label=f"{column} (Rolling Mean)")
plt.xlabel("Year")
plt.ylabel("Subscribers (in millions)")
plt.title("Telecom Subscriber Growth (Smoothed with Rolling Mean)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, orient="h", palette="Set2")
plt.xlabel("Subscribers (in millions)")
plt.title("Distribution of Subscribers per Telecom Company")
plt.show()
```

```python
df['Top_Company'] = df.idxmax(axis=1)
print("\nTop Telecom Company Each Year:\n", df['Top_Company'])
```

```python
plt.figure(figsize=(12, 6))
df.plot(kind='hist', bins=20, alpha=0.6, figsize=(12, 6))
plt.xlabel("Subscribers (in millions)")
plt.title("Subscriber Distribution Across Companies")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=df.index, y=df["Bharti Airtel"], color='red')
plt.xlabel("Year")
plt.ylabel("Subscribers (in millions)")
plt.title("Trend Analysis for company_to_analyze")
plt.show()
```

```python
plt.figure(figsize=(30, 20))
sns.stripplot(data=df, jitter=True, size=7)  
plt.title("Strip Plot of Telecom Subscribers")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.plot(kind='hist', bins=20, alpha=0.6)
plt.xlabel("Subscribers (in millions)")
plt.title("Subscriber Distribution Across Companies")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.iloc[:, 1:].boxplot(rot=90)
plt.title("Boxplot of Subscriber Distributions")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.iloc[:, 1:].hist(bins=10, figsize=(12, 8), layout=(4, 4), edgecolor='black')
plt.suptitle("Histograms of Telecom Subscribers")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.set_index(df.columns[0]).plot(kind='bar', figsize=(12, 6))
plt.title("Bar Chart of Telecom Companies Over Time")
plt.xlabel("Year")
plt.ylabel("Subscribers (millions)")
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.savefig("bar_chart.png")
plt.show()
```

```python
plt.figure(figsize=(20, 5))
sns.violinplot(data=df, inner="quartile")
plt.title("Violin Plot of Features")
plt.xticks(rotation=45)
plt.show()
```
